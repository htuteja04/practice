Pandas have 2 basic data types, series & DataFrames. series can be considered a column & DataFrame a table

```python
names = ['Himanshu', 'Jay', 'Ken', 'Ron']
name_series = pd.Series(names, name='Names')
name_series
```

    -iloc is index based operator
    -loc is label based
    -both take arguments as row first, column second

eg. df.iloc[row_index range, column_index range]

```python
df_spo.iloc[:10, :10] # upto 10th row & 10th column
```

```python
df_spo.iloc[:5, :] #upto 5th row, all columns
```

```python
df_spo.iloc[:, :3] #all rows, upto 3 columns
```

```python
df_spo.iloc[[1,3,5], [1,2,4,6]] #custom rows & columns
```

```python
df_spo.iloc[-5:, -5:] #last 5 columns & last 5 rows
```

loc is strictly column name based

```python
df_spo.loc[:, 'tempo'] #all rows, specific column
```

```python
df_spo.loc[:10, ['artist', 'duration_mins']] #10 rows, specific columns
```

```python
df_spo.loc[(df_spo['artist']=='Rihanna') & (df_spo['year'] == 2008)]
```

```python
df_spo.loc[(df_spo['genre'].isin(['rock', 'pop']))]
```
