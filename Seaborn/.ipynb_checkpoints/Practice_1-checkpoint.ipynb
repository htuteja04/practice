```python
flight = pd.read_csv('Datasets/Viz/flight_delays.csv',index_col='Month')
```

```python
mean_US = flight['US'].mean()
```

```python
flight['US'].fillna(mean_US, inplace=True)
```

```python
plt.figure(figsize=(18,10))
sns.heatmap(flight, annot=True);
```

```python
plt.figure(figsize=(18,10))
sns.heatmap(flight, annot=True, linewidths=0.5);
```
