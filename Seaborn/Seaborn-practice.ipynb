```python
sns.histplot(data=p,x="bill_depth_mm", hue="species", multiple="stack");
```

```python
sns.displot(data=p, x="flipper_length_mm", hue="species", col="species");
```

```python
sns.displot(data=p, x="flipper_length_mm", hue="sex", col="species");
```

```python
sns.scatterplot(data=p, x="flipper_length_mm", y="bill_length_mm", hue="sex");
```

```python
fig, ax = plt.subplots(1,2,figsize=(8,4))
sns.scatterplot(data=p, x="bill_length_mm",y="bill_depth_mm", legend=False, hue="sex", ax=ax[0])
sns.histplot(data=p, x="body_mass_g", hue="sex", ax=ax[1]);
```

```python
fig, ax = plt.subplots(1,2,figsize=(8,4),  gridspec_kw=dict(width_ratios=[4, 3]))
sns.scatterplot(data=p, x="bill_length_mm",y="bill_depth_mm", legend=False, hue="sex", ax=ax[0])
sns.histplot(data=p, x="body_mass_g", hue="sex", ax=ax[1], shrink=0.8, alpha=0.8);
```
